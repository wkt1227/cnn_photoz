```python
from astropy.table import Table
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
table = Table.read('../data/external/COSMOS2015_Laigle+_v1.1.fits.gz')
```

```python
table2 = table[:]['ALPHA_J2000', 'DELTA_J2000', 'FLAG_PETER', 'PHOTOZ', 'TYPE', 'ZPDF', 'ZPDF_L68', 'ZPDF_H68', 'ZMINCHI2', 'CHI2_BEST', 'ZP_2', 'CHI2_2', 'NBFILT', 'ZQ', 'CHIQ', 'MODQ', 'MODS', 'CHIS', 'MODEL', 'AGE', 'EXTINCTION', 'MNUV', 'MU', 'MB', 'MV', 'MR', 'MI', 'MZ', 'MY', 'MJ', 'MH', 'MK', 'MNUV_MR', 'CLASS', 'MASS_MED', 'MASS_MED_MIN68', 'MASS_MED_MAX68', 'MASS_BEST', 'L_NU', 'L_R', 'L_K']
```

```python
table2
```

```python
df = table2.to_pandas()
df
```

```python
# 銀河のみを抽出
df2 = df[(df['PHOTOZ'] == df['ZPDF'])]
df2
```

```python
df3 = df2[(df2['TYPE'] == 0)]
df3
```

```python
# 0 < z < 9
df4 = df3[(df3['ZPDF'] > 0) & (df3['ZPDF'] < 9)]
df4
```

```python
# ZP_2 < 0
df5 = df4[df4['ZP_2'] < 0]
df5
```

```python
df6 = df5[df5['MASS_MED'] > 7.5]
df6
```

```python
df7 = df6[(df6['CHI2_BEST']/df6['NBFILT']) < 5]
df7
```

```python
df8 = df7[df7['CHI2_BEST'] < df7['CHIS']]
df8
```

```python
df9 = df8[(df8['ZPDF']-df8['ZPDF_L68']) < 0.05*(1+df8['ZPDF'])]
df9
```

```python
df10 = df9[(df9['ZPDF_H68']-df9['ZPDF']) < 0.05*(1+df9['ZPDF'])]
df10
```

```python
df10[df10['FLAG_PETER']==0]
```

```python
df10.to_csv('../data/interim/COSMOS2015.csv')
```
