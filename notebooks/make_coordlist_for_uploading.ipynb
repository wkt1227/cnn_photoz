```python
import pandas as pd
```

```python
p_df = pd.read_csv('../data/interim/photoz_object.csv')
s_df = pd.read_csv('../data/external/specz_object.csv.gz')
```

```python
# photoz r-band
f = open('../data/processed/p_coordlist_r.txt', 'w')
f.write('#? rerun        ra       dec       sw    sh   filter  image  mask variance type name\n')
for r in p_df.itertuples():
    objid = r._2
    ra = r.ra
    dec = r.dec
    f.write(f'  pdr2_wide  {ra}  {dec}  10asec  10asec  HSC-R  true   false   false   coadd   {objid}_r\n')
f.close()
```

```python
# specz r-band
f = open('../data/processed/s_coordlist_r.txt', 'w')
f.write('#? rerun        ra       dec       sw    sh   filter  image  mask variance type name\n')
for r in s_df.itertuples():
    objid = r._1
    ra = r.ra
    dec = r.dec
    f.write(f'  pdr2_wide  {ra}  {dec}  10asec  10asec  HSC-R  true   false   false   coadd   {objid}_r\n')
f.close()
```
